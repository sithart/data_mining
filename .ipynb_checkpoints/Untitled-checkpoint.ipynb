```python
# Module import
import os
import re
import random
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
user_agent_list = [
'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.1.1 Safari/605.1.15',
'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:77.0) Gecko/20100101 Firefox/77.0',
'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:77.0) Gecko/20100101 Firefox/77.0',
'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
]
```

```python
url = "https://www.in.pampers.com/pregnancy/baby-names/article/top-100-tamil-baby-girl-names-of-2021"

# Pick a random user-agent
user_agent = random.choice(user_agent_list)
# set the headers
headers = {'User-Agent': user_agent}
# Make the request
response = requests.get(url, headers=headers)
pattern = r'[0-9]+.'
name_array = []
if response.status_code == 200:
    try:
        html = response.content
        soup = BeautifulSoup(html, 'html.parser')
        parent = soup.find("div", attrs={"class":"article-text"})
        for child in parent.find_all('p', attrs={"class":"rich-text-text"}):
            para = child.get_text()
            if len(para) > 400:
                section_arr = para.split('\r\n')
                for arr in section_arr:
                    value = arr.split('-')
                    try:
#                         new_string = re.sub(pattern,'',value[0])
#                         name = new_string
#                         name_meaning = value[1]
                        name_array.append(value)
                        
                    except Exception as e:
                        print(e)
        name_dict = {}
        for arrs in name_array:
            if len(arrs) == 2:
                new_string = re.sub(pattern,'',arrs[0])
                name = new_string
                name_meaning = arrs[1]
                name_dict[name] = name_dict[name_meaning]
            else:
                print(arrs)
    except Exception as e:
        print(e)
else:
    print(response.status_code)
```
